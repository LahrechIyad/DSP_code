## Convolution:

```python
import numpy as np

# Define the unit step function
def unit_step(n):
    return np.where(n >= 0, 1, 0)

# Define delta function (discrete impulse)
def delta(n):
    return np.where(n == 0, 1, 0)

# Function to find positions of delta impulses
def find_delta_positions(signal):
    return np.where(signal != 0)[0]

# Question (a)
n = np.arange(-10, 10)  # Define a range for n
x1 = 2 * delta(n) + 5 * delta(n - 5)  # x1[n]
h1 = unit_step(n + 1) - unit_step(n - 1)  # h1[n]
y1 = np.convolve(x1, h1, mode='full')  # Convolution
n_y1 = np.arange(2 * n[0], 2 * n[-1] + 1)  # Time indices for y1

# Question (b)
x2 = (
    delta(n - 1)
    + 2 * delta(n - 2)
    + 3 * delta(n - 3)
    + 4 * delta(n - 4)
    + 5 * delta(n - 5)
    + 6 * delta(n - 6)
)  # x2[n]
h2 = delta(n + 1) + delta(n) + delta(n - 1) + 2 * delta(n - 2)  # h2[n]
y2 = np.convolve(x2, h2, mode='full')  # Convolution
n_y2 = np.arange(2 * n[0], 2 * n[-1] + 1)  # Time indices for y2

# Question (c)
k_max = 10  # Set a limit for k
x3 = np.sum([k * delta(n - k) for k in range(k_max + 1)], axis=0)  # x3[n]
h3 = delta(n - 2)  # h3[n]
y3 = np.convolve(x3, h3, mode='full')  # Convolution
n_y3 = np.arange(2 * n[0], 2 * n[-1] + 1)  # Time indices for y3

# Output results with delta positions
print("y1[n] (Result of x1[n] * h1[n]):", y1)
print("Delta positions in y1[n]:", n_y1[find_delta_positions(y1)])

print("\ny2[n] (Result of x2[n] * h2[n]):", y2)
print("Delta positions in y2[n]:", n_y2[find_delta_positions(y2)])

print("\ny3[n] (Result of x3[n] * h3[n]):", y3)
print("Delta positions in y3[n]:", n_y3[find_delta_positions(y3)])
```

```python
import numpy as np

# Define delta function (discrete impulse)
def delta(n):
    return np.where(n == 0, 1, 0)

# Function to find positions of delta impulses
def find_delta_positions(signal):
    return np.where(signal != 0)[0]

# Function to parse user input into delta functions
def parse_signal(signal_str, n):
    signal = np.zeros_like(n)
    terms = signal_str.split('+')
    for term in terms:
        term = term.strip()
        if 'delta' in term:
            # Extract the shift in delta(n-...)
            if '(n' in term:
                start = term.index('(n') + 2
                end = term.index(')')
                shift_str = term[start:end].strip()  # Extract the shift
                shift = int(shift_str) if shift_str else 0  # Default to 0 if no shift
            else:
                shift = 0  # Default to 0 if delta() is written without (n)
            signal += delta(n - shift)
        elif term.isdigit():
            # Handle constant value, like "3"
            constant = int(term)
            signal += constant * np.ones_like(n)
        elif '*' in term:  # Handle scalar multiplication like "2*delta(n-3)"
            parts = term.split('*')
            scalar = int(parts[0].strip())
            delta_part = parts[1].strip()
            if 'delta' in delta_part:
                if '(n' in delta_part:
                    start = delta_part.index('(n') + 2
                    end = delta_part.index(')')
                    shift_str = delta_part[start:end].strip()
                    shift = int(shift_str) if shift_str else 0
                else:
                    shift = 0
                signal += scalar * delta(n - shift)
    return signal

# Ask the user for input
print("Please define your signals in the format (e.g., 'delta(n-2) + 2*delta(n-3) + 3'): ")

# Get the first signal (x[n])
x_str = input("Enter x[n]: ")
# Get the second signal (h[n])
h_str = input("Enter h[n]: ")

# Define the range for n (time indices)
n = np.arange(-10, 10)

# Parse the user input to generate the signals
x = parse_signal(x_str, n)
h = parse_signal(h_str, n)

# Perform convolution
y = np.convolve(x, h, mode='full')

# Compute the time indices for the result of the convolution
start_index = n[0] + n[0]  # Convolution starts at n[0] + n[0]
end_index = n[-1] + n[-1]  # Convolution ends at n[-1] + n[-1]
n_y = np.arange(start_index, end_index + 1)

# Correct delta positions
delta_positions = n_y[find_delta_positions(y)]

# Output results
print("\ny[n] (Result of x[n] * h[n]):", y)
print("Delta positions in y[n]:", delta_positions)
```

## Fast Fourrier Transform :

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f1 = 1000  # frequency of the original signal (in Hz)
fs = 8000  # sampling frequency (in Hz)
amplitude = 5  # amplitude of the cosine wave

# Frequency range for plotting (from 0 to 20 kHz)
f_range = np.linspace(0, 20000, 1000)

# Original signal spectrum: two spikes at f1 and -f1
original_spectrum = np.zeros_like(f_range)
original_spectrum[np.abs(f_range - f1) < 10] = amplitude  # Approximate spike at f1
original_spectrum[np.abs(f_range + f1) < 10] = amplitude  # Approximate spike at -f1

# Sampled signal spectrum (replicating original spectrum at multiples of fs)
sampled_spectrum = np.zeros_like(f_range)

# Generate the replicated spectra for multiples of fs
spike_freqs = []
for n in range(-3, 3):  # We only need from -15000 to 17000 Hz
    # Approximate spike at f1 + n*fs and -f1 - n*fs
    sampled_spectrum[np.abs(f_range - (n * fs + f1)) < 10] = amplitude
    sampled_spectrum[np.abs(f_range + (n * fs + f1)) < 10] = amplitude

    # Store the spike frequencies for labeling
    spike_freqs.append(n * fs + f1)
    spike_freqs.append(-n * fs - f1)

# Plotting the spectra
plt.figure(figsize=(10, 6))

# Plot the original signal spectrum
plt.plot(f_range, original_spectrum, label="Original Signal Spectrum", color="blue", lw=2)

# Plot the sampled signal spectrum
plt.plot(f_range, sampled_spectrum, label="Sampled Signal Spectrum (0 to 20 kHz)", color="red", linestyle="--", lw=2)

# Adding text annotations for each spike
for freq in spike_freqs:
    # Only annotate if the frequency is within the plotted range
    if 0 <= freq <= 20000:
        plt.text(freq, amplitude + 0.2, f'{freq} Hz', ha='center', va='bottom', color='black', fontsize=9)

# Formatting the plot
plt.title("Spectrum of Original and Sampled Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.xlim(0, 20000)
plt.ylim(0, 6)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
